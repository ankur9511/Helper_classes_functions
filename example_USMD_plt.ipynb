#### Import Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import USMD_plt as hlp
```

#### Plot PMF

```python
try1 = hlp.US_1D(dname="npt_bias/*",
                 histarr=np.arange(0,31,0.5),
                 eq=200,nprod=2000,temperature=300)

try1.extraction(CVcol=2,CVname='/CV',plumedname='/plumed.dat')

f_i = try1.results['f_i']
df_i = try1.results['df_i']

ymin = min(f_i)
f_i = f_i - ymin
ymin = 0

fig = plt.figure(figsize=(3,3))
ax = fig.add_axes([0,0,1,1])
hlp.axplt(ax,try1.bin_center_i,f_i,
      color='C4',xlim=np.array([]),lw=2,
      llabel=r'sphere r = 6$\rm \AA$',
      xticks=np.array([0,10,20,30]),
      xtickparams="0,10,20,30",
      ylabel=r'$\rm F/kT$',
      xlabel=r'$\rm \~N $',
      ylim=np.array([]),
      showleg="yes")
reference = np.loadtxt("../bulk_water/F_Ntilde_WHAM.out",skiprows=4)
hlp.axplt(ax,reference[:,0],reference[:,1],
      color='k',xlim=np.array([]),lw=0,
      llabel=r'ref sphere r = 6$\rm \AA$',
      xticks=np.array([0,10,20,30]),
      xtickparams="0,10,20,30",
      showleg="yes",markersize=2)

fig = plt.figure(figsize=(3,3))
ax = fig.add_axes([0,0,1,1])
hlp.axplt(ax,try1.bin_center_i,df_i,
      color='C5',xlim=np.array([]),lw=2,
      llabel=r'sphere r = 6$\rm \AA$',
      xticks=np.array([0,10,20,30]),
      xtickparams="0,10,20,30",
      ylabel=r'$\rm Standard deviation$',
      xlabel=r'$\rm \~N $',
      ylim=np.array([]),
      showleg="yes")
```
